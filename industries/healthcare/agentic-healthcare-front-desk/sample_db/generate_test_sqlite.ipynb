```python
# This notebook is for generating the sample sqlite database filled with available appointment times.
# This will be the sample test database that the appointment assistant can connect to.
# The interaction between the appointment assistant's tools and the sample database 
# is for illustration purposes only, to illustrate how we can interact with any type of database / API via API valls.

import sqlite3

import pandas as pd

import numpy as np

def generate_random_pd_df(possible_providers, days = 60, start_hour = 8, amount_work_hours = 9):
    start_date = np.datetime64('today')
    end_date = np.datetime64('today') + days

    list_dates = np.arange(start_date, end_date)
    
    
    list_datetimes = [np.datetime64(str(d)+" "+"08:00:00")+ 3600 * np.random.choice(np.arange(0, amount_work_hours)) for d in list_dates]

    list_doctors =  np.random.choice(possible_providers, len(list_dates))
    
    return list_datetimes, list(list_doctors)


possible_providers = ["Doctor A", "Doctor B", "Doctor C", "Doctor D"]
possible_appt_types = ["Adult physicals",
                       "Pediatric physicals",
                       "Follow-up appointments",
                       "Sick visits",
                       "Flu shots",
                       "Other vaccinations",
                       "Allergy shots",
                       "B12 injections",
                       "Diabetes management",
                       "Hypertension management",
                       "Asthma management",
                       "Chronic pain management",
                       "Initial mental health consultations",
                       "Follow-up mental health appointments",
                       "Therapy sessions",
                       "Blood draws",
                       "Urine tests",
                       "EKGs",
                       "Biopsies",
                       "Medication management",
                       "Wart removals",
                       "Skin tag removals",
                       "Ear wax removals"
                       ]



num_days=365


datetime, doctor, appointment_type, patient = [], [], [], []
for type in possible_appt_types:
    list_datetimes, list_doctors = generate_random_pd_df(days = num_days, possible_providers = possible_providers)
    datetime.extend(list_datetimes)
    doctor.extend(list_doctors)
    appointment_type.extend(list(np.repeat(type, num_days)))
    patient.extend(list(np.repeat(None, num_days)))

print(datetime)
print(doctor)
print(appointment_type)
print(patient)

dict = {'datetime': datetime, 'doctor': doctor, "appointment_type": appointment_type, "patient": patient} 
   
df = pd.DataFrame(dict)
df["datetime"] = pd.to_datetime(df["datetime"])
   
print(df) 
```

```python
import sqlite3
local_file = "test_db_testing.sqlite"
conn = sqlite3.connect(local_file)
cursor = conn.cursor()

df.to_sql("appointment_schedule", conn, if_exists="replace")
conn.commit()
conn.close()
```
